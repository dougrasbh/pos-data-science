```python
import pandas as pd
```

```python
df_raw_data = pd.read_csv('../raw/Olympic Games.csv', encoding='ansi')
df_raw_data
```

```python
len(df_raw_data)
```

```python
df_grouped_by_country = df_raw_data.groupby('Country', as_index=False).size().rename(columns={'size': 'Total Medals'})
df_grouped_by_country = df_grouped_by_country.sort_values(by='Total Medals', ascending=False).reset_index(drop=True) 
df_grouped_by_country
```

```python
df_grouped_by_country.to_csv('processed_olympic_games_by_country.csv', index=False)
```
