```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_theme()
```

```python
score = np.array([16, 13, 15, 12, 10, 20, 25, 40, 30, 25, 60, 50])
plt.plot(score)
plt.show()
```

```python
plt.plot(score, c='r')
plt.show()
```

```python
plt.plot(score, c='black', ls='--')
plt.show()
```

```python
plt.plot(score, c='red', ls='--', marker='s')
plt.show()
```

```python
plt.plot(score, c='red', ls='--', marker='s', ms=18)
plt.show()
```

```python
score_2 = np.array([33, 30, 40, 35, 30, 25, 10, 60, 45, 40, 50, 30])
plt.plot(score, c='r', ls='--', marker='s')
plt.plot(score_2, c='blue', ls='--', marker='s')
plt.show()
```

```python
score_2 = np.array([33, 30, 40, 35, 30, 25, 10, 60, 45, 40, 50, 30])
plt.plot(score, c='r', ls='--', marker='s', label='first')
plt.plot(score_2, c='blue', ls='--', marker='s', label='second')
plt.legend()
plt.show()
```

```python
score_2 = np.array([33, 30, 40, 35, 30, 25, 10, 60, 45, 40, 50, 30])
plt.plot(score, c='r', ls='--', marker='s', label='first')
plt.plot(score_2, c='blue', ls='--', marker='s', label='second')
plt.legend(loc='lower right')
plt.show()
```

```python
score_2 = np.array([33, 30, 40, 35, 30, 25, 10, 60, 45, 40, 50, 30])
plt.plot(score, c='r', ls='--', marker='o', label='first')
plt.plot(score_2, c='blue', ls='--', marker='^', label='second')
plt.legend(loc='lower right')
plt.show()
```

```python
score_3 = np.array([39, 42, 37, 37, 40, 35, 28, 45, 45, 50, 55, 40])
plt.plot(score, c='r', ls='--', marker='o', label='first')
plt.plot(score_2, c='blue', ls='--', marker='^', label='second')
plt.plot(score_3, c='black', marker='s', label='third')
plt.legend(loc='lower right')
plt.show()
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
         'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

plt.plot( meses, score, c='red', ls='--', marker='o', label='Decision Tree')
plt.plot( meses, score_2, marker='^', label='K-NN')
plt.plot( meses, score_3, c='black', marker='s', label='Naive Bayes')
plt.legend(loc='lower right')
plt.title('Avaliao de desempenho de modelos em 2019')
plt.ylabel('Acurcia')
plt.show()
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
         'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

plt.plot( meses, score, c='red', ls='--', marker='o', label='Decision Tree')
plt.plot( meses, score_2, marker='^', label='K-NN')
plt.plot( meses, score_3, c='black', marker='s', label='Naive Bayes')
plt.legend(loc='lower right')
plt.title('Avaliao de desempenho de modelos em 2019')
plt.ylabel('Acurcia')
plt.show()
```

```python
plt.fill_between(meses, score,  label='Decision Tree')
plt.fill_between(meses, score_2, label='K-NN')
plt.fill_between(meses, score_3, label='Naive Bayes')
plt.ylabel('Acurcia')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.fill_between(meses, score,  alpha=0.4, label='Decision Tree')
plt.fill_between(meses, score_2, alpha=0.4, label='K-NN')
plt.fill_between(meses, score_3, alpha=0.4, label='Naive Bayes')
plt.ylabel('Acurcia')
plt.legend(loc='upper left')
plt.plot(meses, score)
plt.plot(meses, score_2)
plt.plot(meses, score_3)
plt.show()
```

```python
models = ['Decision Tree', 'K-NN', 'Naive Bayes']
x = meses
y = [score, score_2, score_3]

# Plot
plt.margins(0, 0)
plt.stackplot(x, y, labels=models, alpha=0.7)
plt.legend(loc='upper left')
plt.title('Grfico de reas empilhadas')
plt.show()
```

```python
# Clculo dos percentuais de cada modelo
total = score + score_2 + score_3
perc1 = score / total
perc2 = score_2 / total
perc3 = score_3 / total

y = [perc1, perc2, perc3]

# Plot
plt.stackplot(x, y, labels=models, alpha=0.7)
plt.margins(0, 0)
plt.title('Grfico de reas 100% empilhadas')
plt.legend(loc='upper left')
plt.show()
```
